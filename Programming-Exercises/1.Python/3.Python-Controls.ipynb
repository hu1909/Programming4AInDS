###1.1.1 if statement

```python
import math
x = int(input("Enter an integer:"))
print(f'x = {x}')
if x >= 0:
  print('x >= 0; SQRT(x) = {:<10.2f}'.format(math.sqrt(x)))
```

###1.1.2 if-else statement


```python
valid_low, valid_high = 0, 10
print(f'valid range: [{valid_low}, {valid_high}]')
x = int(input("Enter an integer:"))
if (x >= valid_low) & (x <= valid_high):
  print(f'x = {x}, a valid value')
else:
  print(f'x = {x}, it is not inside of the valid range')
```

###1.1.3 else-if statement

```python
valid_low, valid_high = 0, 10
print(f'valid range: [{valid_low}, {valid_high}]')
x = int(input("Enter an integer:"))
if x < valid_low:
  print(f'x = {x}, it is smaller than the lower-bound; so, it is invalid')
elif x > valid_high:
  print(f'x = {x}, it is larger than the upper-bound; so, it is invalid')
else:
  print(f'x = {x}, a valid value')
```

```python
import statistics as stats 

"""
Median of threes: A, B, C
Fill your code to:
1. get an integer from user
2. determine the median value
3. make the report as shown in the output cell followed
Note: call stats.median([A, B, C]) to get the median by library 'statistics' for comparison

"""
def median(data):
    new_data = sorted(data)
    n = len(new_data)

    if n % 2 == 1:
        return new_data[n // 2]
    else: 
        return (new_data[n // 2 - 1] + new_data[n // 2]) / 2 

input_list = []
A = int(input("Enter a number:"))
input_list.append(A)
B = int(input("Enter a number:"))
input_list.append(B)
C = int(input("Enter a number:"))
input_list.append(C)
print(f"A, B, C = {A}, {B}, {C}")


print(f"Median of [{A}, {B}, {C}]: {median(input_list)}")
print(f"Median by Statistics Library: {stats.median([A,B,C])}")
```

###2.1.1 for with function range
Notes:
* ***range(start=0, stop, step=1)***: generates a sequence from ***start*** (included)to ***stop***(excluded), increments by ***step***, 



```python
#start is 0 by default, stop is 10 (not included 10), increment=1 by default
for v in range(10): 
  print(v)
```

```python
#start is 2, stop is 10 (not included 10), increment=1 by default
for v in range(2, 10):
  print(v)
```

```python
#start is 2, stop is 10 (not included 10), increment=2

for v in range(2, 10, 2):
  print(v)
```

## 1.1- Question 1: select values using **index**, **list of indices** for **implicit** vs **explicit** index
* Information:
  * ```s``` is series object
    * implicit keys: 0,1,2,...
    * explicit keys: see ```s.index``` or ```s.keys()```
  
* Requirements:
  1. Using syntax as in the output cell:
    * ```s[65]```,
    * ```s.loc[65]```
    * ```s.iloc[0]```
    * and so on, to select values from the series
  2. Show the results as in the output cell

* Guidelines:
  * Notice on **values with the same keys**




````python
# Declare variables
N = 15

np.random.seed(0)
values = np.random.randint(0, 100, (N, ))
keys = np.random.randint(0, 100, (N, ))
keys[4] = keys[0]
s = pd.Series(values, index=keys)

print('Series (s): '); print(s)
print('s.index:  ', s.index)
print('s.keys(): ', s.index)
print('s.values: ', s.values)
print('-'*100)

print("s[65]:")
print(s[65])
print("s.loc[65]:")
print(s.loc[65])
print("s.iloc[65]:")
print(s.iloc[0])
````

- `loc()` - label-based indexing -> Row selection uses the index label <br>
    For example: 
    ```
    df.loc["A"] -> row with index label "A"
    ```
- `iloc()` - position-based indexing -> Row selection uses row position <br>
    For example:
    ```
    df.iloc[0] -> the first row 

    df.iloc[0:3] -> rows at position 0,1,2 
    ```

## 1.2- Question 2: select values using **mask** and **slicing**
* Information:
  * ```s``` is series object
    * implicit keys: 0,1,2,...
    * explicit keys: see ```s.index``` or ```s.keys()```
  
* Requirements:
  1. Using syntax as in the output cell:
    * ```s.loc[65:46]```,
    * ```s.iloc[0:3]```
    * and so on, to select values from the series
  2. Show the results as in the output cell

* Guidelines:
  * Notice: with syntax ```start:stop:step```
    * **implicit**: ```stop``` excluded
    * **explicit**: ```stop``` included

````python
# Declare variables
N = 15

np.random.seed(0)
values = np.random.randint(0, 100, (N, ))
keys = np.random.randint(0, 100, (N, ))
s = pd.Series(values, index=keys)

print('Series (s): '); print(s)
print('s.index:  ', s.index)
print('s.keys(): ', s.index)
print('s.values: ', s.values)
print('-'*100)


print("s.loc[65:46]: (include 46)")
print(s.loc[65:46])
print("s[0:3] (w/o loc or iloc, default: implicit): (exclude 3)")
print(s[0:3])
print("s.iloc[0:3] (exclude 3)")
print(s.iloc[0:3])
print("mask: s >=70:")
print(s >= 70)
print("s[s >= 70]:")
print(s[s >= 70])
````

## 2.11- Categorical columns (object)
* With categorical columns (dtype=object)
  * **need to list unique values**
    * use: ```dataframe[column].unique()```
  * **need to count unique values**
    * use: ```dataframe[column].value_counts()```

````python
# show list of columns that their dtype is object
obj_cols = [v for v in table.columns if table[v].dtype==object]
print("columns with dtype==object", obj_cols)

print("list of country names:")
print(table['country'].unique())

print("list of different values in cca2:")
print(table['cca2'].unique())

print("list of different values in cca3:")
print(table['cca3'].unique())
````

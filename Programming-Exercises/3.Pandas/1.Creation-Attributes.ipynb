# 1- Series


## 1.1- Question 1:
* Information:
  * N is number of integers will be generated by you

* Requirements:
  1. Generate N integers randomly (uniform distribution)
  2. Create a series (named s) from N number generated
  3. Show the results as in the output cell
    * Print series s
    * Print the index and values of s

* Guidelines:
  * To print the index: ```list(s.index)```
  * To print the values: ```list(s.values)```



````python
# Declare variables
N = 15

np.random.seed(42)
random_value = np.random.randint(0, 100, size=N)
print(random_value)
s = pd.Series(random_value)

print("Series (s):")
print(s)
print(f"s.index: {list(s.index)}")
print(f"s.values: {s.values.tolist()}")
````

## 1.2- Question 2:
* Information:
  * N is number of integers will be generated by you

* Requirements:
  1. Generate N integers randomly (uniform distribution), those will be the ```data``` of the series being created.
  2. Generate N integers randomly (uniform distribution), those will be the ```index``` of the series being created.
  3. Create a series (named s) from the data and the index generated
  3. Show the results as in the output cell
    * Print series s
    * Print the index and values of s

* Guidelines:
  * To print the index: ```list(s.index)```
  * To print the values: ```list(s.values)```



````python
# Declare variables
N = 15

np.random.seed(0)

random_data = np.random.randint(0, 100, size=N)
random_index = np.random.randint(0,100, size=N)
s = pd.Series(random_data, random_index)

print("Series (s):")
print(s)

print(f"s.index: {list(s.index)}")
print(f"s.values: {s.values.tolist()}")
````

## 1.3- Question 3:

* Requirements:

  2. Create a series (named s) from the data generated, and the keys is the list of characters stored in ```keys```

  3. Show the results as in the output cell

## 1.4- Question 4:
* Information:

* Requirements:
  1. Show attributes of series ```s```
  2. Call method: ```head```, ```tail``` and ```sample``` to show the first 3, the last 3 and sampled 3 items from series ```s```

## 2.3- Question 3:

* Requirements:
  1. Create a DataFrame (named df) from columns given in ```student```, ```rank```, and ```mark```
  2. Show the result as in the output cell

````python
# Variables
student =  ["Amit", "John", "Jacob", "David", "Steve"]
rank = [1, 4, 3, 5, 2]
mark = [95, 70, 80, 60, 90]

#YOUR CODE IS HERE
student_series = pd.Series(student, name="student")
rank_series = pd.Series(rank, name="rank")
mark_series = pd.Series(mark, name="mark")

df = pd.concat([student_series, rank_series, mark_series], axis=1)

print("DataFrame (df):")
print(df)
````
